```python
import csv 

f = open('train.csv', encoding='utf8') # 또는 'eur-kr', 'cp949'
data = csv.reader(f)
next(data)
data = list(data)
```

```python
print(len(data))
```

```python
for row in data :
    if row[4]=='':
        row[4]='0'
    print(row[4])
```

```python
for row in data :
    row[3] = float(row[3])
    row[4] = float(row[4])
    row[-1] = float(row[-1])
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')

for row in data :    
    plt.scatter(row[3],row[-1], color = 'hotpink')    
plt.xlabel('평균 기온(°C)') # x축 이름
plt.ylabel('교통사고 발생량') # y축 이름
plt.show()
```

평균 기온은 교통사고 발생량과 큰 관련을 갖지 않는다.
교통사고가 가장 많이 발생한 날의 평균 기온은 약 10°C였다.
교통사고가 가장 적게 발생한 날의 평균 기온은 약 -3°C였다. 

```python
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
a = []
b = []
c = []
d = []
e = []
f = []
g = []
h = []
i = []
j = []

for row in data :
    if -15<row[3] and row[3] <= -10 :
        a.append(row[-1])
    if -10<row[3] and row[3] <= -5 :
        b.append(row[-1])
    if -5<row[3] and row[3] <= 0 :
        c.append(row[-1])
    if 0<row[3] and row[3] <= 5 :
        d.append(row[-1])
    if 5<row[3] and row[3] <= 10 :
        e.append(row[-1])
    if 10<row[3] and row[3] <= 15 :
        f.append(row[-1])
    if 15<row[3] and row[3] <= 20 :
        g.append(row[-1])
    if 20<row[3] and row[3] <= 25 :
        h.append(row[-1])
    if 25<row[3] and row[3] <= 30 :
        i.append(row[-1])
    if 30<row[3] and row[3] <= 35 :
        j.append(row[-1])
    
    
    
plt.boxplot([a,b,c,d,e,f,g,h,i,j],labels = ['-15°C~-10°C','-10°C~-5°C','-5°C~0°C','0°C~5°C','5°C~10°C','10°C~15°C','15°C~20°C','20°C~25°C','25°C~-30°C','30°C~35°C'],showmeans = True)
plt.xlabel('기온(°C)') # x축 이름
plt.xticks(rotation=45)
plt.ylabel('평균 교통사고 발생량') # y축 이름
plt.show()
```

기온에 따라 교통사고 발생량의 평균은 크게 변하지 않았다. 평균적으로 하루에 약 100건의 교통사고가 발생하였다. 

```python
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
for row in data :    
    if row[4] != 0 :
        plt.scatter(row[4],row[-1],color = 'purple')
plt.xlabel('비온 날 강수량(mm)') # x축 이름
plt.ylabel('교통사고 발생량') # y축 이름
plt.show()
    
```

강수량은 교통사고 발생량과 큰 관련을 갖지 않는다.
비 오는 날 중 교통사고가 가장 많이 발생한 날의 강수량은 약 20mm였다.
비 오는 날 중 교통사고가 가장 적게 발생한 날의 강수량은 약 5mm였다.

```python
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
a = []
b = []
c = []
d = []
e = []
f = []
g = []
h = []
i = []
j = []
k = []
l = []

for row in data :    
    if 0<row[4] and row[4] <= 5 :
        a.append(row[-1])
    if 5<row[4] and row[4] <= 10 :
        b.append(row[-1])
    if 10<row[4] and row[4] <= 15 :
        c.append(row[-1])
    if 15<row[4] and row[4] <= 20 :
        d.append(row[-1])
    if 20<row[4] and row[4] <= 30 :
        e.append(row[-1])
    if 30<row[4] and row[4] <= 40 :
        f.append(row[-1])
    if 40<row[4] and row[4] <= 60 :
        g.append(row[-1])
    if 60<row[4] and row[4] <= 80 :
        h.append(row[-1])
    if 80<row[4] and row[4] <= 100 :
        i.append(row[-1])
    if 100<row[4] and row[4] <= 120 :
        j.append(row[-1])
    if 120<row[4] and row[4] <= 140 :
        k.append(row[-1])
    if 140<row[4] :
        l.append(row[-1])
        
plt.boxplot([a,b,c,d,e,f,g,h,i,j,k,l],labels = ['0mm~5mm','5mm~10mm','10mm~15mm','15mm~20mm','20mm~30mm','30mm~40mm','40mm~60mm','60mm~80mm','80mm~100mm','100mm~120mm','120mm~140mm','140mm~'],showmeans = True)
   
    
plt.xlabel('비온 날 강수량(mm)') # x축 이름
plt.xticks(rotation=45)
plt.ylabel('평균 교통사고 발생량') # y축 이름
plt.show()
```

강수량이 60mm ~ 80mm 일 때 평균 교통사고 발생량이 가장 많았다. 

결론 : 
기온이나 강수량에 따라서 교통사고 발생량이 유의미한 변화를 보이지는 않았다. 아무래도 여러 복합적인 요인들이 영향을 주고 있기 때문에 변인통제가 제대로 이루어지지 않은 한계가 있는 것 같다. 

```python
for row in data:
    row[2]=row[2].split('-')
    print(row[2])
```

```python
for row in data:
    row[-1]=int(row[-1])
    print(row[-1])
```

```python
for row in data:
    row[2][1]=int(row[2][1])
    print(row[2][1])
```

```python
ac = [0 for i in range(4)]
print(ac)
for row in data:
    if row[2][1] in [3,4,5]:
        ac[0]+=row[-1]
    elif row[2][1] in [6,7,8]:
        ac[1]+=row[-1]
    elif row[2][1] in [9,10,11]:
        ac[2]+=row[-1]
    elif row[2][1] in [12,1,2]:
        ac[3]+=row[-1]
print(ac)
for i in range(0,4):
    ac[i]=int(ac[i])
    ac[i]-=20000
print(ac)
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
s=['봄','여름','가을','겨울']
plt.rc('font', family='Malgun Gothic')
plt.title('계절별 교통사고발생량(2017~2019)') # 그래프 제목
plt.bar(range(len(ac)), ac, width=0.6)
plt.xticks(range(len(ac)),s)
plt.ylabel('20000건+')
plt.show()
```

비교가 어려워 20000건을 빼주었다
겨울이 제일 많을 줄 알았으나 가을이 제일 많았다. 그다음으로는 여름, 봄, 겨울 순이다.

```python
for row in data:
    if row[8]=='':
        row[8]=100000
    else:
        row[8]=float(row[8])
        row[8]=round(row[8])
    print(row[8])
    print(type(row[8]))
```

```python
acs = [0 for i in range(9)]
a=0
b=0
c=0
d=0
e=0
f=0
g=0
h=0
i=0
print(acs)
for row in data:
    if row[8]==1:
        acs[0]+=row[-1]
        a=a+1
    elif row[8]==2:
        acs[1]+=row[-1]
        b=b+1
    elif row[8]==3:
        acs[2]+=row[-1]
        c=c+1
    elif row[8]==4:
        acs[3]+=row[-1]
        d=d+1
    elif row[8]==5:
        acs[4]+=row[-1]
        e=e+1
    elif row[8]==6:
        acs[5]+=row[-1]
        f=f+1
    elif row[8]==7:
        acs[6]+=row[-1]
        g=g+1
    elif row[8]==8:
        acs[7]+=row[-1]
        h=h+1
    elif row[8]==9:
        acs[8]+=row[-1]
        i=i+1
acs[0]=acs[0]/a
acs[1]=acs[1]/b
acs[2]=acs[2]/c
acs[3]=acs[3]/d
acs[4]=acs[4]/e
acs[5]=acs[5]/f
acs[6]=acs[6]/g
#acs[7]=acs[7]/h
acs[8]=acs[8]/i
print(acs)
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
s=['1','2','3','4','5','6','7','8','9']
plt.rc('font', family='Malgun Gothic')
plt.title('일 최심적설(반올림)별 교통사고의 평균(2017~2019)') # 그래프 제목
plt.bar(range(len(acs)), acs, width=0.6)
plt.xticks(range(len(acs)),s)
plt.show()
```

일 최심적설 -> 하루 중에 눈이 가장 많이 쌓였을 때의 깊이
눈이 많이 쌓일 수록 교통사고량이 많을 줄 알았으나 교통사고에 큰 영향을 주지 않았다

```python
import csv

f = open('train.csv')
data = csv.reader(f)
next(data)
result =[]

for row in data:
    if row[-3] != '':
        result.append(float(row[-3]))
print(result)
print(len(result))
```

```python
for row in data :
    row[-3] = float(row[-3])
    row[-1] = float(row[-1])
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic') 

for row in data :
    plt.scatter(row[-3],row[-1], color= 'hotpink')
plt.xlabel('평균 전운량')
plt.ylabel('교통사고 발생량')
plt.show()
```

산점도로 나타내었을 때, 전운량에 따른 교통사고 발생량이 많았는지, 적었는지 알지 못했다.

```python
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
a=[]
b=[]
c=[]
for row in data :
    if 0<row[-3] and row[-3]<4 :
        a.append(row[-1])
    if 4<row[-3] and row[-3]<8 :
        b.append(row[-1])
    if 8<row[-3] and row[-3]<10 :
        c.append(row[-1])
        
plt.boxplot([a,b,c], labels = ['0~4','4~8','8~10'], showmeans=True)
plt.xlabel('평균 전운량')
plt.ylabel('교통사고 발생량')
plt.show()
```

산점도로 나타내었을 때, 전운량에 따른 교통사고 발생량이 많았는지, 적었는지 알지 못했다.
전운량에 따른 교통사고를 boxplot으로 나타내보니, 의외의 결과로 전운량이 0~4 일 때, 교통사고 발생량이 많았다.
전운량이 8~10 일 때, 교통사고 발생량의 최댓값인 140보다 큰 이상치가 3개가 나왔다.

따라서 전운량에 따른 교통사고 발생량은 전운량이 적을 때 더 많은 사고가 난 것으로 변인통제가 이루어지지 않은 한계가 있는 것 같다.

```python
for row in data :
    if row[6] == '':
        row[6] = 10000
    row[6] = float(row[6])
    print(row[6])
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic') 

for row in data :
    if row[6] != 10000 :
        plt.scatter(row[6],row[-1], color = 'green')
plt.xlabel('일사량')
plt.ylabel('교통사고 발생량')
plt.show()
```

산점도로 나타내었을 때, 일사량에 따른 교통사고 발생량이 많았는지, 적었는지 알지 못했다.

```python
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
a=[]
b=[]
c=[]
for row in data :
    if 0<row[6] and row[6]<10 :
        a.append(row[-1])
    if 10<row[6] and row[6]<20 :
        b.append(row[-1])
    if 20<row[6] and row[6]<30 :
        c.append(row[-1])
        
plt.boxplot([a,b,c], labels = ['0~10','10~20','20~30'], showmeans=True)
plt.xlabel('일사량')
plt.ylabel('교통사고 발생량')
plt.show()
```

일사량에 따른 교통사고를 boxplot으로 나타내보니, 의외의 결과로 일사량이 0~10 일 때, 교통사고 발생량이 많았다.

따라서 일사량에 따른 교통사고 발생량은 일사량이 적을 때 더 많은 사고가 난 것으로 변인통제가 이루어지지 않은 한계가 있는 것 같다.
